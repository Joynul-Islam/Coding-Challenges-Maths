```python
#Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
%matplotlib inline
plt.style.use('seaborn')
```

```python
#dictonary to store probabilitesof each card being draw on first play
probDictAvg = {"escape": 0,"blackmail": 0, "c_c": 0, "dayoff": 0, "payrise": 0, "securitybreach": 0, "bosscompliment": 0, "deadline": 0, "employerOftheMonth": 0}

#number of samples to take
sampleSize = 100

for i in range(sampleSize):
    #normal cards
    deck = ["blackmail","blackmail","blackmail","blackmail","blackmail","Colleague-Clash","Colleague-Clash","Colleague-Clash","Colleague-Clash","day-Off","day-Off","day-Off","day-Off","day-Off","pay-rise","pay-rise","pay-rise","pay-rise","pay-rise","security-breach","security-breach","security-breach","security-breach","security-breach","boss compliment","boss compliment","boss compliment","boss compliment","boss compliment","deadline","deadline","deadline","deadline","Employer of the Month"]
    #charecter cards
    characterCards = ["Gossip","glamour","kick ass","brat","slob","workaholic","vlogger","Nerd","Beachbum","who?","office couple","short-fuse"]
    player1 = []
    player2 = []
    player3 = []
    player4 = []
    
    #deck and characters are shuffled
    random.shuffle(deck)
    random.shuffle(characterCards)

    #Each player will get 2 normal cards
    for i in range(2):
        player1.append(deck.pop())
    for i in range(2):
        player2.append(deck.pop())
    for i in range(2):
        player3.append(deck.pop())
    for i in range(2):
        player4.append(deck.pop())
    
    #each playre gets an escape card
    player1.append("escape")
    player2.append("escape")
    player3.append("escape")
    player4.append("escape")
    
    #each player gets a charecter card
    player1.append(characterCards.pop())
    player2.append(characterCards.pop())
    player3.append(characterCards.pop())
    player4.append(characterCards.pop())
    
    #Loser and extra escape carded placed back in deck
    for i in range(2):
        deck.append("escape")
    for i in range(3):
        deck.append("You're Fired!")
    
    #collect data
    probDict = {"escape": 0,"blackmail": 0, "c_c": 0, "dayoff": 0, "payrise": 0, "securitybreach": 0, "bosscompliment": 0, "deadline": 0, "employerOftheMonth": 0}
    for i in deck:
        if i == "escape":
            probDict["escape"] += 1
        elif i == "blackmail":
            probDict["blackmail"] += 1
        elif i == "Colleague-Clash":
            probDict["c_c"] += 1
        elif i == "day-Off":
            probDict["dayoff"] += 1
        elif i == "pay-rise":
            probDict["payrise"] += 1
        elif i == "security-breach":
            probDict["securitybreach"] += 1
        elif i == "boss compliment":
            probDict["bosscompliment"] += 1
        elif i == "deadline":
            probDict["deadline"] += 1
        elif i == "Employer of the Month":
            probDict["employerOftheMonth"] += 1
    
    probDictAvg["escape"] += probDict["escape"]
    probDictAvg["blackmail"] += probDict["blackmail"]
    probDictAvg["c_c"] += probDict["c_c"]
    probDictAvg["dayoff"] += probDict["dayoff"]
    probDictAvg["payrise"] += probDict["payrise"]
    probDictAvg["securitybreach"] += probDict["securitybreach"]
    probDictAvg["bosscompliment"] += probDict["bosscompliment"]
    probDictAvg["deadline"] += probDict["deadline"]
    probDictAvg["employerOftheMonth"] += probDict["employerOftheMonth"] 
    
#find mean value
for i in probDictAvg:
    probDictAvg[i] = "{0:.2f}".format((probDictAvg[i]/sampleSize)/29)
 
print(probDictAvg)
```
